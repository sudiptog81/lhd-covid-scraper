```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = [
    'hospital-beds',
    'non-covid-icu-beds',
    'ventilators'
]

for d in datasets:
    df = pd.read_csv(f'data/{d}.csv')
    df[['name', 'label']].groupby(by=['label']).count().plot(kind='pie', y='name', title=d)
```
